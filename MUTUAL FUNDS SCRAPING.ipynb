# MUTUAL FUND INDIA

```python
# pip install requests==2.20.0
# <b>https://www.mutualfundindia.com/MF/return/TopFunds?id=3</b>

# Our aim will be :-
# 1. Retrieving HTML data from the specified website
# 2. Parsing that data for target information
# 3. Storing the target information

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

Step 1: Import the necessary libraries and get the html of the page.
Use urllib.request module to open URLs.
The Beautiful Soup package is used to extract data from html files. The Beautiful Soup library’s name is bs4 which stands for Beautiful Soup, version 4.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

Step 2: Create a Beautiful Soup object from the html.
Beautiful Soup package is used to take the raw html text and break it into Python objects (Parsing the data). It helps formatting and organizing the messy web by fixing bad HTML and presenting us with easily traversable Python objects.

There are two things to note in the below code. 1) we are converting the html data to a beautiful soup object. 2) lxml is a very useful XML/HTML processing library.

```python
html=urlopen("https://www.mutualfundindia.com/MF/return/TopFunds?id=3")
bsobj=BeautifulSoup(html,'lxml')
print(bsobj)
```

Step3 : Store the target information. Our target is to take each row of data and store it in a csv file.
1. Retrieve the table rows

```python
table_rows= bsobj.findAll('tr')
print(table_rows)
```

But, what we need is all table rows in a list so that we can convert the list into data frame.

```python
for row in table_rows:
    each_row=row.findAll("td")
    print(each_row)
```

Note: You are doing two things here. 1) removing html tags using BeautifulSoup and extracting only text 2) Creating an empty list and appending each row of text to the empty list.

```python
lists_of_rows=[]
for row in table_rows:
    each_row=row.findAll("td")
    str_row =str(each_row)
    row_text=BeautifulSoup(str_row,"lxml").get_text()
    lists_of_rows.append(row_text)
print(lists_of_rows)
```

Step4:- The next step is to convert the list into a data frame and get the view of the rows using Pandas

```python
import pandas as pd
import numpy as np
data=pd.DataFrame(lists_of_rows[5:])
print(data)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data
```

```python
data.to_csv("MF.csv",  encoding='utf-8',index=False)
```

# Advisor Khoj

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html=urlopen("https://www.advisorkhoj.com/mutual-funds-research/latest-Mutual-Funds-NAV-52-week-high-nav?category=All&amc=All&option=Latest%20NAV%20vs%2052%20Week%20Highest%20NAV")
bsobj=BeautifulSoup(html,'lxml')
print(bsobj)
```

```python
table_rows= bsobj.findAll('tr')
print(table_rows)
```

```python
for row in table_rows:
    each_row=row.findAll("td")
    print(each_row)
```

```python
lists_of_rows=[]
for row in table_rows:
    each_row=row.findAll("td")
    str_row =str(each_row)
    row_text=BeautifulSoup(str_row,"lxml").get_text()
    lists_of_rows.append(row_text)
print(lists_of_rows)
```

```python
import pandas as pd
import numpy as np
data=pd.DataFrame(lists_of_rows[5:])
print(data)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data
```

```python
data.to_csv("AdvisorKhoj.csv",  encoding='utf-8',index=False)
```
